```python
import matplotlib.pyplot as plt 
import numpy as np 

```

```python
def getInfoFromFile(fileName, x_list, y_list):
    if not x_list == False:
        x_list.clear()
    if not y_list == False:
        y_list.clear()    
    roster = list()
    with open(fileName, 'r') as file:
        lines = [line.rstrip('\n') for line in file]
        roster = lines[0::2];
        for element in roster:
            x_list.append(float(element))
        roster = lines[1::2]
        for element in roster:
            y_list.append(float(element))    
```

```python
x_first_version = list()
y_first_version = list()

x_second_version = list()
y_second_version = list()

file1 = 'DistributionBernuolli.txt'
file2 = 'DistributionMonteCarlo.txt'
```

```python
getInfoFromFile(file1, x_first_version, y_first_version)
getInfoFromFile(file2, x_second_version, y_second_version)
```

```python

x = np.arange(len(x_first_version))

width = max(y_first_version)
plt.figure(figsize=(20,5))
ax = plt.subplot()
rect1 = ax.bar(x-0.2, y_first_version, width=0.4, label="Bernuolli", color='blue')
rect2 = ax.bar(x+0.2, y_second_version, width=0.4, label="MonteCarlo", color='orange')
ax.set_ylabel('Distributions')
ax.set_xlabel('Factors')
ax.set_xticks(x)
ax.set_xticklabels(x_first_version)
ax.set_title('Distibutions when different factors')

def autolabel(rects):
     for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rect1)
autolabel(rect2)
ax.legend()
plt.show
```

```python
import math
differenceDistribution = list()

for firstDistr, secondDistr in zip(y_first_version, y_second_version):
    differenceDistribution.append(math.fabs(firstDistr-secondDistr))
    

```

```python
x = np.arange(len(x_first_version))
plt.figure(figsize=(20,7))
ax = plt.subplot()
rect1 = ax.bar(x, differenceDistribution, width=0.4, color='blue')
ax.set_ylabel('Difference')
ax.set_xlabel('Factors')
ax.set_xticks(x)
ax.set_xticklabels(x_first_version)
ax.set_title('Difference Bernuolli and MonteCarlo methods')
autolabel(rect1)
ax.legend()
plt.show
```
