```python
import matplotlib.pyplot as plt 
```

```python
def getInfoFromFile(fileName, x_list, y_list):
    roster = list()
    with open(fileName, 'r') as file:
        lines = [line.rstrip('\n') for line in file]
        for line in lines:
            if(line.isdigit()):
                roster.append(int(line))
        x_list += roster[0::2]
        y_list += roster[1::2]        
```

```python
x_first_version = list()
y_first_version = list()

x_second_version = list()
y_second_version = list()
```

```python
file1 = 'InfoBernuolli.txt'
file2 = 'InfoMonteCarlo.txt'
```

```python
getInfoFromFile(file1, x_first_version, y_first_version)
getInfoFromFile(file2, x_second_version, y_second_version)
```

```python
fig = plt
react1 = plt.plot(x_first_version, y_first_version, color='red', label='MonteCarlo')
react2 = plt.plot(x_second_version, y_second_version,  color='blue', label='Bernuolli')
plt.title('Distribution of 10000 M')
plt.xlabel('N')
plt.ylabel('Distr')
plt.legend()
```
